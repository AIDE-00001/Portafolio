## Misión 1: Equipo de Ventas

Necesitamos construir una visualización para indicar los **10 principales clientes por total de ventas**.

Con ello, el equipo de ventas puede pensar en estrategias para mantener a estos clientes comprando en nuestra tienda.

```python
#crear un nuevo data frame donde queremos agrupa(por el nombre de cliente unicamente la columna de
# ventas)sumar las ventas.quiero los 10 mas relevante.generar una copia
clientes = ventas.groupby(['nombre_cliente'])['ventas'].sum().nlargest(10).copy()
#para que no quede como una serie resetear el indice
clientes = clientes.reset_index()
#vizualisamos
clientes
```

```python
# dic = {'nombre_cliente':'Clientes', 'ventas':'Ventas'}-->*
# clientes = clientes.rename(columns=dic)-->
# * es otra foma de hacerlo

#asignamos la lista (colmna cliente,columna ventas)-->ambas deben tenner las mismas filas de lo contrario debemos crear un diccioonario primero
clientes.columns = ['Clientes','Ventas']
#para el ranki podemostomar los clientes y sumar 1
clientes['Ranking'] = clientes.index + 1
#dejar que el ranking sea como el indice parametro inplace=true
clientes.set_index('Ranking', inplace = True)
#visualizamos
clientes
```

```python
#style es un obejeto que puede cambiar el formato:letra,fuente,color,si es cursiva...
estilo = clientes.style
estilo
```

```python
#cambiamos el fomato(llave ventas:el formato que quiero)
estilo.format({'Ventas':'{:,.2f}'})
estilo
```

# Misión 2: Equipo Comercial

Necesitamos construir una visualización que permita asociar el **total de ventas y el total de lucros por tipo de producto**.

Ello permitirá que el equipo comercial pueda enfocarse en estrategias para apalancar los ingresos que se reciben con cada tipo de producto vendido en la tienda.

```python
#crear un nuevo data frame agrupoamos(tipo de producto)tomamos la culumna ventas,ganancias. suma
#cuando hay doble corchete creamos el data frame
ganancias_ventas = ventas.groupby(['tipo_producto'])[['ventas','ganancia']].sum()
ganancias_ventas
```

```python
#asignamos al nombre del indice =tipo de producto
ganancias_ventas.index.name = 'tipo de producto'
ganancias_ventas
```

```python
#crear nuestro estilo_producto= aplicar la propiedad style a gnacias ventas
estilo_producto = ganancias_ventas.style
estilo_producto
```

```python
#Vamos a darle un formato(primero dejarlo en 2 cifras decimales,separar ,para separaralo miles)que muestre USD
estilo_producto.format('USD {:,.2f}')
```

```python
#Vamos a utilizar para resartal el valor maximo y el valor minimo
#.higlight_max es resaltar(color).higlith_min resaltar(red)
#lo almacenamos
estilo_producto = estilo_producto.format('USD {:,.2f}').highlight_max(color='lightgreen').highlight_min(color='red')
estilo_producto
```

```python
#tecnica para meojorar la presentacion usando un a escala de colores
#para se intesigique cuando se mas  se vende y menos color con lo que menos se venda
#esto lo podemos acer con el metodo bakgroup(colocar el mapa de colores en este caso verde)
estilo_producto.background_gradient(cmap='Greens')
```

```python
#trabajaremos con css se parece a los diccinarios de python
#objeto llamado encabezado{debmos estrablecer lo paramtros selector (a que le vamos hacer la modificaion) y propiedades que vamos alterar en el  }
#props:colocar las propiedades respetando cada un de las categorias que tiene
encabezado = {
              'selector':'th',
              'props':'font-weight:bold; font-family:Arial; text-align:center; text-transform:capitalize'
}
```

```python
#vamos aplicarlo a nuesto estilo(estilo,overwrite=false es para que no se pierda las tranformaciones que hemos realizado)
estilo_producto.set_table_styles([encabezado],overwrite=False)
```

# Misión 3: Equipo de Logística

Necesitamos construir una visualización que nos permita entender la **distribución de pedidos por región en Argentina**.

Así, el equipo de logística va a saber dónde hay más oportunidad de enfocar los esfuerzos de material y mano de obra.

```python
#preparando nuestrosdatos
#creamos un data frame(vamos hacer un conteo de ventas por regiones)
region = pd.DataFrame(ventas['region'].value_counts())
#vamos a modificar el nombre de la columna y le asignamos el nomre
region.columns = ['N° Pedidos']
#vamos a cambiar el nombre del indice
region.index.name = 'Región'
#guardamos en una variable
region
```

```python
#importamos la libreria numpy nos ayuda de forma mas eficiente con numeros
import numpy as np
```

```python
#Crear variable (almcenamos los porcentajes)convertila en un to_numpy
porcentaje = region['N° Pedidos'].to_numpy()
#multiplicamos por 100 dividido entre la suma
porcentaje = 100*porcentaje/porcentaje.sum()
#visualizar el porcntaje
porcentaje
```

```python
#crear nueva columna la almacenamos en porcentaje
region['Porcentaje de Pedidos'] = porcentaje
#visualizamos
region
```

```python
#creacion de nuestro estilo
estilo_region = region.style
#visualizamos
estilo_region
```

```python
#vamos a aliniear los encabezados y que nuestra tabla sea todo en blaco utilizando css
encabezado = {
              'selector':'th',
              'props':'font-weight:bold; font-family:Arial; text-align:right; background-color:white'
}
#props:peso de la fuente:bold,familia de fuente arial,aliniacion a la derecha,color de bacgrund=blanco
#para los datos(celdas)
#en selector en lugae de th va ser td=table data
celdas = {
              'selector':'td',
              'props':'background-color:white'
}

estilo_region.set_table_styles([encabezado,celdas])
```

```python
#vamos a darle formato unicamente columna porcentaje pedidos
estilo_region.format({'Porcentaje de Pedidos':'{:.2f} %'})\
                    .bar(subset='Porcentaje de Pedidos',vmin=0,vmax=100,color='#9cd3bb')
```

# Misión 4: Equipo de Logística

Nuevamente el equipo de logística desea replantear su estructura y organización de provisionamiento de productos e, para ello, se pensó en una visualización que permita identificar el estándar de la **cantidad de productos solicitados por mes por departamento**.

```python
#creamos un nuevo data frame que es copia de ventas
meses = ventas.copy()
#organizar el data frame por la fecha de pedido
meses = meses.sort_values('fecha_pedido')
#visualizamos
meses
```

```python
#tomar el data frame creamos el atribo mes vamo atomar de meses la columna fecha pdido.strtime(formato de string para tiempo)
meses['mes'] = meses['fecha_pedido'].dt.strftime('%Y - %b')
#vamos a quitar el idice
meses = meses.reset_index(drop=True)
#visulizamos
meses
```

```python
#crear una tabla dinamica es ecificando (que podamos ver todas las columnas)
pd.set_option('display.max_columns',None)
```

```python
#pivot table es una tabla dinamica(idice van hacer los departamentos, columnas van se la columna mes, los valores va a ser la columna cantidad, columna de agregacion va ser lla suma,paraque lo deje tal cual con sort=false)
ventas_mensuales = meses.pivot_table(index=meses.departamento, columns='mes', values='cantidad', aggfunc='sum', sort=False)
#visualizamos
ventas_mensuales
```

```python
#Fijar los nombres de lo que representa cada uno de los indices
#creams nuestro estilo tomamos nuestro data frame ventas _mesuales
estilo_mensual = ventas_mensuales.style
#visualizamos
estilo_mensual
```

```python
#vamos a dejarlo fijo(que queremos dejar fijo axis=index)
estilo_mensual.set_sticky(axis='index')
```

```python
#mejorar el formato de la tabla
#Creamos nuestros dicionarios
#{solo quermos la columna, propiedades }
columnas = {
            'selector':'.col_heading',
            'props':'font-weight:normal; font-family:Arial'
}
#establecer para la tabla{tabla,datos, propiedades la alineacion}
tabla = {
            'selector':'td,th',
            'props':'text-align:left'
}
#vamos a tabjar en los indices{index_name,propiedades alinear a la derecha..}
indices = {
            'selector':'.index_name',
            'props':'font-weight:bold; font-family:Arial; text-align:right'
}
#tomar nestros estilo mensual([]overwrite=false para no perder lo que hemos realizado)
estilo_mensual.set_table_styles([columnas,tabla,indices], overwrite=False)
```

# Misión 5: Informe de Desempeño

El informe de desempeño es un documento muy importante para la empresa. En este, todos los que trabajan en la tienda pueden entender cómo está la empresa. Debido a ello, las tablas bien construídas son esenciales para mostrar los datos de la mejor forma.

Por lo tanto, necesitamos construir una visualización que permita entender la **relación entre tipos de clientes y modo de envío de productos de acuerdo con las ventas**. De esta forma, será posible destacar en el informe cuál es el modo de envío y cuáles son los que más generaron lucro en las ventas.

```python
#Vamos a tomar nuevo data frame(vamo a tener clientes, tipo de eenvio,los valores van a ser las ventas, la funcion de agregacion va se la suma)
df_cliente = ventas.pivot_table(index='segmento_cliente',columns='modo_envio',values='ventas',aggfunc='sum')
#visualizamos
df_cliente
```

```python
#Necesitamos los valores totale(creamos dos nuevos atributos)
# es igul a la suma de las columnas
df_cliente['Total'] = df_cliente.sum(axis=1)
#crear la columna en el indice, hacemos la suma
df_cliente.loc['Total'] = df_cliente.sum()
#visualizamos
df_cliente
```

```python
#creamos un estilo, formato a toda la tabla
compra_cliente = df_cliente.style.format('{:,.2f}')
#visualizamos
compra_cliente
```

```python
#crar nuestro diccionario tabla y le colocamos los parameros que deseamos, not(.index_name)es para que no se lo agrege a los indice
tabla={
    'selector':'td,th:not(.index_name)',
    'props':'font-weight:normal; font-family:Arial; text-align:center: background-color:white'
}
#crear diccionario para los indices
indice={
    'selector':'.index_name',
    'props':'font-weight:normal; font-style:italic; text-align:right: color:#696969'
}
#tomamos nuestro estilo
compra_cliente.set_table_styles([tabla,indice])
```

```python
#vamos aplicar el metodo st_table_styles sobre nuestro objeto estilo
compra_cliente.set_table_styles({
    'Total':[
        {
            'selector':'th',
            'props':'border-top: 2px solid #181818'
        },
        {
            'selector':'td',
            'props':'border-top: 2px solid #181818'
        }
    ],
    'B2B':[
        {
            'selector':'th',
            'props':'border-top: 2px solid #181818'
        },
        {
            'selector':'td',
            'props':'border-top: 2px solid #181818'
        }
    ]
}, overwrite=False, axis=1)
```

```python
#visualizar la tabla que creamos en el codigo que sigue
colores_columna = pd.DataFrame(['false','true','false'], index=df_cliente['Total'].index, columns=['Total'])
colores_columna
```

```python
#crea un selector personalizado
compra_cliente.set_table_styles({
    'Total':[{
      #donde sea verdadero lo haga
        'selector':'.true',
        'props':'background-color: #D8D8D8'
    }]
},overwrite=False,axis=0)

#construir un data frame que va hacer nuestra mascara ([valores que va tener flaso,verdadero,falso para que tenga las mimas dimensiones],indice va el df del cliente,columna va ser total)
colores_columna = pd.DataFrame(['false','true','false'], index=df_cliente['Total'].index, columns=['Total'])

#otro metodo que podemos utilizar
compra_cliente.set_td_classes(colores_columna)
```
