## 2. Ventas por categoría

---
**Descripción**
En este debes calcular la cantidad de productos vendidos por categoría en cada tienda. La idea es agrupar los datos por categoría y contar el número de ventas de cada tipo, mostrando las categorías más populares de cada tienda.



```python
cant_categoria_tienda1 = tienda1.groupby('Categoría del Producto')['Producto'].count().sort_values(ascending=False)
cant_categoria_tienda1
```

```python
cant_categoria_tienda2 = tienda2.groupby('Categoría del Producto')['Producto'].count().sort_values(ascending=False)
cant_categoria_tienda2
```

```python
cant_categoria_tienda3 = tienda3.groupby('Categoría del Producto')['Producto'].count().sort_values(ascending=False)
cant_categoria_tienda3
```

```python
cant_categoria_tienda4 = tienda4.groupby('Categoría del Producto')['Producto'].count().sort_values(ascending=False)
cant_categoria_tienda4
```

```python
categorias_mas_popular_de_cada_tienda = {
    'Tienda 1': cant_categoria_tienda1.idxmax(),
    'Tienda 2': cant_categoria_tienda2.idxmax(),
    'Tienda 3': cant_categoria_tienda3.idxmax(),
    'Tienda 4': cant_categoria_tienda4.idxmax()
}
categorias_mas_popular_de_cada_tienda
```

```python
import matplotlib.pyplot as plt

tiendas = [tienda1, tienda2, tienda3, tienda4]

fig, axes = plt.subplots(2, 2, figsize=(14,10))

for i, tienda in enumerate(tiendas):

    productos = tienda.groupby('Producto')['Producto'].count().sort_values(ascending=True).head(5)

    ax = axes[i//2, i%2]
    productos.plot(kind='barh', ax=ax)

    ax.set_title(f'Tienda {i+1}')
    ax.set_xlabel('Cantidad de ventas')
    ax.set_ylabel('Producto')

fig.suptitle('Productos más y menos vendidos de cada tienda', fontsize=16)

plt.tight_layout()
plt.show()
```

# 5. Envío promedio por tienda
---
**Descripción**
En este paso, debes calcular el costo de envío promedio para cada tienda. El objetivo es comprender cuánto se gasta, en promedio, en el envío de cada tienda.  Sabiendo que el cliente paga el envío.

```python
cost_envio_prom_para_cada_tienda = {
    'Tienda 1': tienda1['Costo de envío'].mean(),
    'Tienda 2': tienda2['Costo de envío'].mean(),
    'Tienda 3': tienda3['Costo de envío'].mean(),
    'Tienda 4': tienda4['Costo de envío'].mean()
}

# To print the dictionary with values formatted to two decimal places:
for tienda, costo in cost_envio_prom_para_cada_tienda.items():
    print(f"{tienda}: {costo:.2f}")
```

## Costo de envio promedio por tienda

```python
import matplotlib.pyplot as plt

# Calcular costo de envío promedio
costos_envio = [
    tienda1['Costo de envío'].mean(),
    tienda2['Costo de envío'].mean(),
    tienda3['Costo de envío'].mean(),
    tienda4['Costo de envío'].mean()
]

tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

plt.figure()

plt.plot(tiendas, costos_envio, marker='o')

plt.title('Costo de envío promedio por tienda')
plt.xlabel('Tiendas')
plt.ylabel('Costo de envío promedio')

plt.show()
```
