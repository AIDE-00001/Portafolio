# Visualización de datos con Python: Matplotlib, Seaborn y Plotly

```python
import pandas as pd

df = pd.read_csv('/content/inmigrantes_canada.csv')
df.head()
```

```python
#Obtener en tupla cuamtas filas y columnas tiene
df.shape
```

```python
#Para tener una version geeneral del data frame que tenemos
df.info()
```

```python
#Vamos a configurar como indice el pais
df.set_index('Pais',inplace=True)
df
```

```python
#Generar un data frame que tenga en la columna los años en una listas
anos = list(map(str,range(1980,2014)))
anos
```

```python
#Aplicar el filtro con el metodo.loc del pais colombia dentro del filtro que creamos anos
colombia = df.loc['Colombia',anos]
colombia
```

```python
#Creamos un data frame con la informacion anterior con un diccionario
#colocamos el año:el objeto colobia tomamos los indices crear una lista con la cantidad
#de inmigantes de colombia
col_dict = {'Año':colombia.index.tolist(),
            'Inmigrantes':colombia.values.tolist()}
col_dict
```

```python
#Crear un data frame
#pedmos los ultimos 5 registros
datos_col = pd.DataFrame(col_dict)
datos_col.tail()
```

## Matplotlib
https://matplotlib.org/

```python
import matplotlib.pyplot as plt
```

```python
#figure() ajustamos el tamaño con el parametro figize (va en pulgadas)
#Puedo hacer un plot(gráfica) voy a dibujar año en eje x y imigrantes en eje y
#xticks se refiere a yaitas que aparencen abajo y le doy un rango de años
#Para evitar que aparecsca donde esta en el espacio de memoria con plt.show()
plt.figure(figsize=(8,4))
plt.plot(datos_col['Año'],datos_col['Inmigrantes'])
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'])
plt.show()
```

```python
#plt.title(Colocamos el titulo)
#plt.xlabel(Etiquete para el je x que va ser año)
#plt.ylabel(la etiqueta del eje y numero de emigrantes)
#plt.show()debe ser el ultimo comando
plt.figure(figsize=(8,4))
plt.plot(datos_col['Año'],datos_col['Inmigrantes'])
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'])
plt.title('Inmigración de colombianos hacia Canadá')
plt.xlabel('Año')
plt.ylabel('Número de Inmigrantes')
plt.show()
```

```python
#Vamos a generar las variable que vano a tener el lienzo fig y ax que es la grafica
#plt.subplots que es donde puedo definir mi figura(grafica)
#le coloco ax.plot()...nos parece una grafica la diferecia es que ya lo guardamos en una variable
#utlizado la funcionn de forma dinamica para ello
#ax.xaxis.set_mejor_locator (utiliza un localizador multiple(de 5 en 5 vaya colocando mi tick))
#ax.set_title() con el comando set _title se coloca el tituño c0n \n coloco 1l periodo 1980-2013
#ax.set_xlabel() rotulo en x
#ax.set_ylabel() el rotulo de y
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(datos_col['Año'],datos_col['Inmigrantes'])
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.set_title('Inmigración de colombianos hacia Canadá\n1980 - 2013')
ax.set_xlabel('Año')
ax.set_ylabel('Número de Inmigrantes')
plt.show()
```

```python
#Gráfico de caja para refenciar diversos valores descriptivos de estadistica descriptiva
#Vamos a trabajar con dos figura
#en subplot(puedo colocar las fila y columnas con la dimension de la figura)
# axs[0] nos dice que es la grafica de la izquierda -->grafico de tendencia
#se colocan los datos parecidos  la grafica anterior
#axs.xaxis.set_major_locator()para colocar el rango de los ticks
#axs.grid()-->Con ello colocamos una cuadricula a grafica
#Realizamos lo mismo para la sigiuiente grafica solo cambia axs[1] para que vaya izquierda
#axs[1].boxplot -->grafico de caja
#plt.show()para mostrar la grafica

fig, axs = plt.subplots(1,2,figsize=(10,3))

axs[0].plot(datos_col['Año'],datos_col['Inmigrantes'])
axs[0].set_title('Inmigración de colombianos hacia Canadá\n1980 - 2013')
axs[0].set_xlabel('Año')
axs[0].set_ylabel('Número de Inmigrantes')
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].grid()

axs[1].boxplot(datos_col['Inmigrantes'])
axs[1].set_title('Boxplot de la inmigración de colombianos\nhacia Canadá 1980 - 2013')
axs[1].set_xlabel('Colombia')
axs[1].set_ylabel('Número de Inmigrantes')
axs[1].grid()

plt.show()
```

```python
#Metodo de pandas que me da la estadistica descriptiva de mis datos
datos_col.describe()
```

```python
#Vamos analizar otros paises

#fig.suptitle(Titulo superior de las cuatro graficas)

#creamos 2 filas y 2 columnas para que podamos colocar 4 gráficas
#axs[0,0] va la grafica de colombia le agregamos lo que queremos que lleve nustra grafica
#axs[0,1] va la grafica de Brasil le agregamos lo que queremos que lleve nustra grafica
#axs[1,0] va la grafica de Argentina le agregamos lo que queremos que lleve nustra grafica
#axs[0,0] va la grafica de Perú le agregamos lo que queremos que lleve nustra grafica
#Cambiamos el ajuste de mis graficos  fig.subplot_adjust(hspace=espacio altura) wspace(el ancho)

#para colocar los datos en la misma escala estableemos un valor minimo ymin=0 y un valor maximo ymax=7000
#Iteramos cada una de ellas con for ravel(trae un arreglo udimensional a uno bidemensional)


#como las graficas tienen los mismos ejes podemos usar un lazo for ax in axs.falt que me permite iterar en cada una de la graficas
#Colocamos en el cuerpo lo que queremos
fig, axs = plt.subplots(2,2,figsize=(10,6))
fig.subplots_adjust(hspace=0.5,wspace=0.3)
fig.suptitle('Inmigración de los 4 mayores países sudamericanos\nhacia Canadá desde 1980 hasta 2013')

axs[0,0].plot(df.loc['Colombia',anos])
axs[0,0].set_title('Colombia')

axs[0,1].plot(df.loc['Brasil',anos])
axs[0,1].set_title('Brasil')

axs[1,0].plot(df.loc['Argentina',anos])
axs[1,0].set_title('Argentina')

axs[1,1].plot(df.loc['Perú',anos])
axs[1,1].set_title('Perú')

ymin = 0
ymax = 7000

for ax in axs.ravel():
  ax.set_ylim(ymin,ymax)

for ax in axs.flat:
  ax.xaxis.set_major_locator(plt.MultipleLocator(5))
  ax.set_xlabel('Año')
  ax.set_ylabel('Número de Inmigrantes')
  ax.grid()

plt.show()
```

```python
#Podemos cambiar el tamaño de la fuente fontsize=18
#cambiar el color de la lineas de la grafica color='g' verde y lw=2.5 seria el grosor
#marker='°' para que tenga puntitos--->solo esta comentado
#loc='left le decimos que coloque a la izquierda nuestro titulo
#las tick le cambiamos los parametros que son cada uno de la rayitas
#ax.grid(linestyle='--') al colocarle ---para que no quede solido
fig, ax = plt.subplots(figsize=(8,3))
ax.plot(datos_col['Año'],datos_col['Inmigrantes'],lw=2.5,color='g') ,#marker='o')
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.set_title('Inmigración de colombianos hacia Canadá\n1980 - 2013', fontsize=18, loc='left')
ax.set_xlabel('Año',fontsize=14)
ax.set_ylabel('Número de Inmigrantes',fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
ax.grid(linestyle='--')
plt.show()
```

```python
#Como nuestro enfoque es colombia vamos a darle un color diferente
#Donde el color de colombia es verde y el de los demas paise es vgris

colores =[]
for pais in sudamerica_sorted.index:
  if pais == 'Colombia':
    colores.append('forestgreen')
  else:
    colores.append('silver')

fig, ax = plt.subplots(figsize=(10,3))

#ax.set_frame on no va tener el recuadro alrededor
ax.set_frame_on(False)
#para eliminar la parte del eje x los numeros que tenia la grafica anterior
ax.xaxis.set_visible(False)
#nosotros estamos redusciendo la longitud del tick por lo que no va aparecer
ax.tick_params(axis='y',which='both',size=0)

ax.barh(sudamerica_sorted.index, sudamerica_sorted['Total'],color=colores)
ax.set_title('Sudamérica: Colombia es el segundo país con más migrantes\nhacia Canadá durante el periodo 1980 - 2013',
             loc='left',fontsize=16)
ax.set_ylabel('')
ax.yaxis.set_tick_params(labelsize=10)
# ax.set_xlabel('Número de Inmigrantes', fontsize=13)-->los comentamos porque ya no los necesitamo
# ax.xaxis.set_tick_params(labelsize=12)-->tambiem lo quitamos

#Vamos a enumerar nestro data frame sudamerica especialmente el valor total
#ax.text(la posicion en x mas 20 pixeles la posicion en y seria i darlerle color negro
#tamaño de la fuente con fontsize, la notacion aliniada a la izquierda,va= va de forma vertical)
for i,j in enumerate(sudamerica_sorted['Total']):
  ax.text(j+20,i,str(j),color='black',fontsize=9,ha='left',va='center')

#Guardamos nuestra imagen para la podamos descargar
fig.savefig('/content/inmigrantes_sudamericanos_canada.png',
            transparent=False,dpi=300,bbox_inches='tight')

plt.show()
```

## Seaborn
https://seaborn.pydata.org/

```python
import seaborn as sns
```

```python
#Configurar un tema por defecto
#Todas las graficas que haremos va a tener un mismo tema
sns.set_theme()
```

```python
#Datos globales historico de inmigracion de canada
#Almacenamos todos los paises, en orden dencendente de menor a mayor, head(10)es el top 10 de los paises
top_10 = df.sort_values('Total',ascending=False).head(10)
top_10
```

```python
#Graficas en seaborn
#barplot(datos, ejex lo que queremos, eje y el nombre de los paise,orient es la horientacion
#en este caso 'h' horizontal)
ax = sns.barplot(data=top_10,x='Total',y=top_10.index,orient='h')
#vmos a congigurar el titulo,configuramos xlabel,ylabel='ao'
ax.set(title='Top 10 países con mayor migración hacia Canadá\n1980 - 2013',
       xlabel='Número de Inmigrantes',
       ylabel='Año')
#para que no me muestre el espacio de memoria
plt.show()
```

```python
#Nosotros podemos utilizr en conuto matplotlib y seaborn
#crear la figur(nuestro lienzoo y nuestra figura y el tamaño de la figura)
fig, ax = plt.subplots(figsize=(8,4))

#la variable ax (almacenamos data frame,nombre el valor total en x,en y el top 10 de lo paises)
ax = sns.barplot(data=top_10,x='Total',y=top_10.index,orient='h')
#el titulo
ax.set_title('Top 10 países con mayor migración hacia Canadá\n1980 - 2013',
             loc='left',fontsize=16)
#Colocar los label
ax.set_xlabel('Número de Inmigrantes',fontsize=14)
#dejamos un espacio vacio en x
ax.set_ylabel('')
#para que no muestre la memoria de almacenamiento
plt.show()
```

```python
#Definir la funcion(generar un parametro)
def generar_grafico(palette):
  fig, ax = plt.subplots(figsize=(8,4))
  ax.set_frame_on(False)
  ax.xaxis.set_visible(False)
  ax.tick_params(axis='y',which='both',size=0)
  #parametro de pallete para que podamos ver los diversas paletas de colores
  ax = sns.barplot(data=top_10,x='Total',y= top_10.index,hue=top_10.index,orient='h',
                   palette=palette, legend=False)
  ax.set_title('Top 10 países con mayor migración hacia Canadá\n1980 - 2013',
              loc='left',fontsize=16)
  ax.set_xlabel('Número de Inmigrantes',fontsize=14)
  ax.set_ylabel('')
  for i,j in enumerate(top_10['Total']):
    ax.text(j+20,i,str(j),color='black',fontsize=9,ha='left',va='center')
  plt.show()
```

```python
#vamos a ver la paleta de azules('Blue_r')-->para una paleta reversa
#las variables de una misma paleta de un mismo color es para variables numericas
generar_grafico('Blues_r')
```

```python
#utilizamos una paleta de valor rocket
generar_grafico('rocket')
```

```python
#probar otra paleta
#las paleatas cuando va de una totalidad a otra(es cuando las variaciones se presenta con respecto a un valor cetral)
generar_grafico('coolwarm')
```

```python
#para este tipo de grafica es mejo una paleta para valores categoricos
generar_grafico('Paired')
```

```python
#podemos usar otra paleta
generar_grafico('tab10')
```

```python
#Podemos cambiar el tema
#colocar la marcaciones
#ajustamos para que funcioe conseaborn
def generar_grafico(palette):
  fig, ax = plt.subplots(figsize=(8,4))
  ax.set(xticklabels=[])
  ax = sns.barplot(data=top_10,x='Total',y= top_10.index,hue=top_10.index,orient='h',
                   palette=palette, legend=False)
  ax.set_title('Top 10 países con mayor migración hacia Canadá\n1980 - 2013',
              loc='left',fontsize=16)
  ax.set_xlabel('Número de Inmigrantes',fontsize=14)
  ax.set_ylabel('')
  sns.despine()
  for i,j in enumerate(top_10['Total']):
    ax.text(j+20,i,str(j),color='black',fontsize=9,ha='left',va='center')
  plt.show()
```

```python
#sns.set_theme es el fondo por defecto
sns.set_theme()
generar_grafico('tab10')
```

```python
#pODEMOS CAMBIAR EL FONDO ESPECIFICANDO EL ESTILO
sns.set_theme(style='darkgrid')
generar_grafico('tab10')
```

```python
#Podemos ver otro estilo
sns.set_theme(style='whitegrid')
generar_grafico('tab10')
```

## Plotly
https://plotly.com/python/

```python
import plotly.express as px
```

```python
#Para que nuestros graficos sean interactivos con la biblioteca plotly

#grafico de linea (nombre de x,nombre de y, titulo de la grafica)
fig = px.line(datos_col,x='Año',y='Inmigrantes',
              title='Inmigración de colombianos hacia Canadá entre 1980 y 2013')
#la funcion update_traces para lalinea(color verder,line_with=ancho de la linea)
fig.update_traces(line_color='green',line_width=3)
#la funcion update de actualizar(ancho,altura(mejorar los ticks con rotacion de -45),
#                nombres de cada uno de los ejes en xanño,yaxisnumero de migrates,
#                font family es la fuente arial,tamaño 14,fondo color gris)
fig.update_layout(width=800,height=400, xaxis={'tickangle':-45},
                  xaxis_title='Año',yaxis_title='Número de Inmigrantes',
                  font_family='Arial',font_size=14, font_color='grey')
#para que no se vea el espacio de memoria
fig.show()
```
